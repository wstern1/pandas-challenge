## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
age_df = purchase_data[['Age Range', 'SN']]
ten_df = age_df.loc[age_df["Age Range"] == "<10", :]
true_ten_df = ten_df.drop_duplicates(subset = 'SN')
ten_total = len(true_ten_df)
fourteen_df = age_df.loc[age_df["Age Range"] == "10-14", :]
true_fourteen_df = fourteen_df.drop_duplicates(subset = 'SN')
fourteen_total = len(true_fourteen_df)
nineteen_df = age_df.loc[age_df["Age Range"] == "15-19", :]
true_nineteen_df = nineteen_df.drop_duplicates(subset = 'SN')
nineteen_total = len(true_nineteen_df)
twentyfour_df = age_df.loc[age_df["Age Range"] == "20-24", :]
true_twentyfour_df = twentyfour_df.drop_duplicates(subset = 'SN')
twentyfour_total = len(true_twentyfour_df)
twentynine_df = age_df.loc[age_df["Age Range"] == "25-29", :]
true_twentynine_df = twentynine_df.drop_duplicates(subset = 'SN')
twentynine_total = len(true_twentynine_df)
thirty_df = age_df.loc[age_df["Age Range"] == "30-34", :]
true_thirty_df = thirty_df.drop_duplicates(subset = 'SN')
thirty_total = len(true_thirty_df)
thirtyfive_df = age_df.loc[age_df["Age Range"] == "35-39", :]
true_thirtyfive_df = thirtyfive_df.drop_duplicates(subset = 'SN')
thirtyfive_total = len(true_thirtyfive_df)
forty_df = age_df.loc[age_df["Age Range"] == "40+", :]
true_forty_df = forty_df.drop_duplicates(subset = 'SN')
forty_total = len(true_forty_df)

age_demo = pd.DataFrame({
    "Total Count" : [ten_total, fourteen_total, nineteen_total, twentyfour_total, twentynine_total, thirty_total, 
                     thirtyfive_total, forty_total],
    "Percentage of Players" : [(ten_total/int(unique_players))*100, (fourteen_total/int(unique_players))*100, 
                               (nineteen_total/int(unique_players))*100, (twentyfour_total/int(unique_players))*100,
                               (twentynine_total/int(unique_players))*100, (thirty_total/int(unique_players))*100,
                               (thirtyfive_total/int(unique_players))*100, (forty_total/int(unique_players))*100]},
    index = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])
age_demo["Percentage of Players"] = age_demo["Percentage of Players"].map(
    "{0:,.2f}%".format)
age_demo.index.name = 'Age Range'
age_demo
```
